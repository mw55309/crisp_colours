```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib import patheffects

typeface='Helvetica Neue'
mpl.rcParams['font.weight']=300
mpl.rcParams['axes.labelweight']=300
mpl.rcParams['font.family']=typeface
mpl.rcParams['font.size']=22

crisp_colours={'SaltVingear':'#6CBEEB',
               'CheeseOnion':'#60C955',
               'Dill':'#5D9E57',
               'Beef':'#9E926F',
               'Chicken':'#EBF25C',
               'Gammon':'#BA4747',
               'TomatoSauce':'#F24E4E',
               'SaltShake':'#F5F5F5',
               'WorcesterSauce':'#B070DB'}

fig,ax = plt.subplots(figsize=(10,10),facecolor='w')

for f,flavour in enumerate(crisp_colours.keys()):
    effects=[patheffects.Stroke(linewidth=4, foreground='k'),
                 patheffects.Stroke(linewidth=1, foreground='w')] ## black text, white outline
    ax.text(1.0,f+0.5,flavour,size=40,ha='center',va='center',path_effects=effects)
    ax.add_patch(plt.Rectangle((0.0,f),3,1,facecolor=crisp_colours[flavour],edgecolor='none'))

ax.set_ylim(0.0,len(crisp_colours))
ax.set_xlim(0,2)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.tick_params(size=0,labelsize=0)

plt.show()
```
